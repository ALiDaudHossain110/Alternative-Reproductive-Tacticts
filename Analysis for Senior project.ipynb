```python
def combine_data(data, data2):
    # Get the highest generation number in data
    last_generation_data = max(data.keys())
    
    # Shift data2 generation numbers to start from the end of data
    shifted_data2 = {
        gen + last_generation_data + 1: value for gen, value in data2.items()
    }
    
    # Combine data and shifted_data2
    combined_data = {**data, **shifted_data2}
    
    return combined_data

# Example usage
```
